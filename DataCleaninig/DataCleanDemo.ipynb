# Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the data

Code

```python
# Load the data (replace 'data.xlsx' with the path to your Excel file)
file_path = "Data.xlsx"
df = pd.read_excel(file_path)
```

# Data Cleaning

Code

```python
# Remove duplicates based on selective column
df = df.drop_duplicates(subset=['Name', 'Age', 'Gender', 'Grade', 'Score', 'Address'])

# Convert columns to appropriate data types
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Score'] = pd.to_numeric(df['Score'], errors='coerce')

# Drop rows with missing or invalid data
df = df.dropna()

# Display the cleaned DataFrame
print("Cleaned Data:")

# Display the cleaned table normally
print(df)

# Display the cleaned table in a tabular format
"""from tabulate import tabulate

print("Cleaned Data in Tabular Format:")
print(tabulate(df, headers='keys', tablefmt='grid'))"""


# Save the cleaned data to a new Excel or CSV file
df.to_csv("cleaned_data.csv", index=False)  # Save as CSV
df.to_excel("cleaned_data.xlsx", index=False)  # Save as Excel
```

# Analysis and Visualization

Code

```python
# Group data by 'Grade' and calculate average score
grade_avg_score = df.groupby('Grade')['Score'].mean()

# Plotting the average scores by grade
plt.figure(figsize=(8, 5))
grade_avg_score.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Scores by Grade', fontsize=14)
plt.xlabel('Grade', fontsize=12)
plt.ylabel('Average Score', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save and show the plot
plt.savefig("average_scores_by_grade.png")
plt.show()

# Additional Visualization: Distribution of students by address
address_counts = df['Address'].value_counts()

plt.figure(figsize=(8, 5))
address_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Student Distribution by Address', fontsize=14)
plt.ylabel('')  # Hide y-label for pie chart
plt.tight_layout()

# Save and show the plot
plt.savefig("student_distribution_by_address.png")
plt.show()
```
